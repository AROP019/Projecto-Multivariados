#Imputación


```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

def imputar_knn(df, k=5):
    df = df.copy()

    date_col = df.columns[0]
    fechas = df[date_col].copy()
    datos = df.drop(columns=[date_col]).copy()

    cols_num = datos.select_dtypes(include=[np.number]).columns
    cols_imputar = [c for c in cols_num if datos[c].isna().mean() < 0.4]

    if not cols_imputar:
        return df, {}

    valores_antes = datos[cols_imputar].copy()
    datos_a_escalar = datos[cols_imputar].copy()


    scaler = StandardScaler()
    datos_escalados = scaler.fit_transform(datos_a_escalar)
    datos_a_imputar_df = pd.DataFrame(datos_escalados, columns=cols_imputar, index=datos_a_escalar.index)

    imputer = KNNImputer(n_neighbors=k)
    datos_imputados_escalados = imputer.fit_transform(datos_a_imputar_df)

    datos_imputados_df = pd.DataFrame(datos_imputados_escalados, columns=cols_imputar, index=datos_a_imputar_df.index)
    datos_finales_desescalados = scaler.inverse_transform(datos_imputados_df)

    datos_knn = datos.copy()
    datos_knn[cols_imputar] = datos_finales_desescalados

    df_final = pd.concat([fechas.reset_index(drop=True), datos_knn.reset_index(drop=True)], axis=1)

    comparacion = {}
    for col in cols_imputar:
        mask_na = valores_antes[col].isna()
        comparacion[col] = pd.DataFrame({
            "Antes (NA)": valores_antes[col][mask_na],
            "Después": datos_knn[col][mask_na]
        })

    return df_final, comparacion
```

```python
dataframes_definitivos = {}
comparaciones_knn = {}

for nombre, df in dataframes.items():
    print(f"\nProcesando {nombre} con KNN...")

    df_imputado, comparacion = imputar_knn(df)

    nuevo_nombre = f"{nombre}Definitivo"
    dataframes_definitivos[nuevo_nombre] = df_imputado
    comparaciones_knn[nombre] = comparacion

print("\nImputación finalizada. Dataframes guardados en dataframes_definitivos.")
```

```python


def extraer_anio(nombre):
    match = re.search(r"(20\d{2})", nombre)
    return match.group(1) if match else None


dataframes_por_anio = {}

for nombre, df in dataframes_definitivos.items():
    anio = extraer_anio(nombre)
    if anio is None:
        print(f"No se encontró año en: {nombre}")
        continue

    if anio not in dataframes_por_anio:
        dataframes_por_anio[anio] = {}

    dataframes_por_anio[anio][nombre] = df


carpeta = "ExcelsPorAnio"
os.makedirs(carpeta, exist_ok=True)

for anio, dfs in dataframes_por_anio.items():
    filepath = f"{carpeta}/{anio}.xlsx"

    with ExcelWriter(filepath, engine="openpyxl") as writer:
        for nombre_df, df in dfs.items():
            sheet_name = nombre_df[:31]
            df.to_excel(writer, sheet_name=sheet_name, index=False)

    print(f"Guardado: {filepath}")
```
